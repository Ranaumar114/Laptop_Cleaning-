```python
laptops.columns = laptops.columns.str.strip().str.lower().str.replace('(','').str.replace(')','')\
.str.replace(' ','_').str.replace('operating_system','os')
```

```python
laptops.info()
```

```python
laptops['screen_size'] = laptops['screen_size'].str.strip('"').astype(float)
```

```python
laptops['screen_size']
```

```python
laptops.info()
```

```python
laptops['screen_size'].mean()
```

```python
laptops.rename({'screen_size' : "screen_size_inch"}, axis= 1 , inplace= True)
```

```python
laptops
```

```python
laptops['cpu_manufacture'] = laptops['cpu'].str.split(n=1, expand= True)[0]
```

```python
laptops
```

```python
laptops['cpu_manufacture'].value_counts()
```

```python
laptops['gpu_manufacture'] = laptops['gpu'].str.split(n=1 , expand= True)[0]
```

```python
laptops
```

```python
laptops['gpu_manufacture'].value_counts()
```

```python
laptops['screen_size_inch'].value_counts()
```

```python
laptops['os'].value_counts()
```

```python
laptops['os'] = laptops['os'].str.replace('Mac OS','macOS')
```

```python
laptops['os'].value_counts()
```

```python
laptops['weight'].value_counts()
```

```python
laptops['weight'] = laptops['weight'].str.replace('kg','').str.replace('kgs','').str.replace('s','').astype(float)
```

```python
laptops['weight'].mean()
```

```python
laptops.rename({'weight' : 'weight_kg'} , axis=1 , inplace= True)
```

```python
laptops
```

```python
laptops.info()
```

```python
laptops['price_euros'] = laptops['price_euros'].str.replace(',','').astype(float)
```

```python
laptops['price_euros'].mean()
```

```python
laptops['os_version'].value_counts()
```

```python
laptops.loc[laptops['os_version'].isnull() , 'os_version'] = 'Unknown_Version'
```

```python
laptops['os_version'].value_counts()
```

```python
laptops['os_version'] = laptops['os_version'].str.replace(' ','')
```
